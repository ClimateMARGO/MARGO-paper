```julia
ADDICE09 = Dict("α1" => 0.0012, "α2" => 0.0023, "α3" => 2.32)
```

```julia
DICE99 = Dict("α1" => -0.0045, "α2" => 0.0035, "α3" => 2)
MARGOv0 = Dict("α1" => 0, "α2" => 0.0033, "α3" => 2)
```

```julia
D(T, p) = p["α1"].*T + p["α2"].*T.^p["α3"]
```

```julia
T = 0.:0.1:5.
```

```julia
using PyPlot
```

```julia
plot(T, D(T, DICE99), label="DICE '99 (Residual)")
plot(T, D(T, ADDICE09), label="AD-DICE '09 (Gross)")
plot(T, D(T, MARGOv0), label="MARGO (Gross) / DICE '99 without linear term")
title("Climate damage function")
ylabel("Climate damages [% GWP]")
xlabel("Global warming [ºC]")
grid(true)
legend()
xlim(0, 5)
```
